```python
## Task 2, Question 2 Code Here
import re
import numpy as np
import pandas as pd

def extract_bathroom_info(text):
    if pd.isna(text):
        return np.nan, np.nan

    text = text.lower().strip()

    # Extract number (including halves)
    number_match = re.search(r'(\d+(\.\d+)?|half)', text)
    if number_match:
        number = number_match.group(1)
        number = 0.5 if number == 'half' else float(number)
    else:
        number = 1  # default if no number found

    # Extract bathroom type; assign NaN if not found
    if 'private' in text:
        bath_type = 'private'
    elif 'shared' in text:
        bath_type = 'shared'
    else:
        bath_type = np.nan

    return number, bath_type

# Apply function to create two new columns in train and test
airbnb_train[['bathroom_qty', 'bathroom_type']] = airbnb_train['bathrooms'].apply(
    lambda x: pd.Series(extract_bathroom_info(x))
)

airbnb_test[['bathroom_qty', 'bathroom_type']] = airbnb_test['bathrooms'].apply(
    lambda x: pd.Series(extract_bathroom_info(x))
)
```

```python
# host_verifications
# Step 1: Function to clean and split the stringified list
def parse_verifications(value):
    methods = []
    if pd.notnull(value):
        value = value.strip("[]").replace("'", "").replace('"', '')
        items = value.split(",")
        for item in items:
            method = item.strip()
            if method:
                methods.append(method)
    return methods

# Step 2: Build a set of all unique methods
unique_methods = set()
for val in airbnb_train['host_verifications']:
    for method in parse_verifications(val):
        unique_methods.add(method)
        
for val in airbnb_test['host_verifications']:
    for method in parse_verifications(val):
        unique_methods.add(method)

# Step 3: Create binary columns for each verification method
for method in sorted(unique_methods):
    col_name = f"has_{method}"
    airbnb_train[col_name] = 0  # Initialize with 0

# Step 3: Create binary columns for each verification method
for method in sorted(unique_methods):
    col_name = f"has_{method}"
    airbnb_test[col_name] = 0  # Initialize with 0
    
# Step 4: Fill the binary columns
for idx, val in airbnb_train['host_verifications'].items():
    methods = parse_verifications(val)
    for method in methods:
        col_name = f"has_{method}"
        airbnb_train.at[idx, col_name] = 1
        
# Step 4: Fill the binary columns
for idx, val in airbnb_test['host_verifications'].items():
    methods = parse_verifications(val)
    for method in methods:
        col_name = f"has_{method}"
        airbnb_test.at[idx, col_name] = 1
```

```python
#========================================================REVIEW========================================================================#
# --------- Step 1: Compute quantile thresholds ---------
p75 = airbnb_train['log_price'].quantile(0.75)
p25 = airbnb_train['log_price'].quantile(0.25)

high_price_listings = airbnb_train[airbnb_train['log_price'] > p75]
low_price_listings = airbnb_train[airbnb_train['log_price'] < p25]

# --------- Step 2: Parse amenities string ---------
def parse_amenities(value):
    value = value.strip("{}").replace('"', "").replace("'", "")
    return [item.strip() for item in value.split(",") if item.strip()]

# --------- Step 3: Count amenity frequency (manual approach) ---------
def get_top_amenities(df, n=10):
    freq_dict = {}
    for val in df['amenities'].dropna():
        amenities = parse_amenities(val)
        for item in amenities:
            if item not in freq_dict:
                freq_dict[item] = 1
            else:
                freq_dict[item] += 1
    # Convert to list of top `n` by frequency
    sorted_items = sorted(freq_dict.items(), key=lambda x: x[1], reverse=True)
    return [item[0] for item in sorted_items[:n]]

top_high_amenities = get_top_amenities(high_price_listings, n=10)
top_low_amenities = get_top_amenities(low_price_listings, n=10)

# --------- Step 4: Add dummy features ---------
def add_amenity_features(df, high_amenities, low_amenities):
    # Create columns
    for amenity in high_amenities:
        df[f'has_high_{amenity}'] = 0
    for amenity in low_amenities:
        df[f'has_low_{amenity}'] = 0
    df['count_high_amenities'] = 0
    df['count_low_amenities'] = 0

    # Fill binary and count values
    for idx, val in df['amenities'].items():
        if pd.notnull(val):
            parsed = parse_amenities(val)
            for amenity in high_amenities:
                if amenity in parsed:
                    df.at[idx, f'has_high_{amenity}'] = 1
                    df.at[idx, 'count_high_amenities'] += 1
            for amenity in low_amenities:
                if amenity in parsed:
                    df.at[idx, f'has_low_{amenity}'] = 1
                    df.at[idx, 'count_low_amenities'] += 1

# --------- Step 5: Apply to both datasets ---------
add_amenity_features(airbnb_train, top_high_amenities, top_low_amenities)
add_amenity_features(airbnb_test, top_high_amenities, top_low_amenities)
```

**Task 2, Question 3**: Impute the missing values for all features in both the training and test datasets.

```python
import pandas as pd
import numpy as np
from scipy.stats import skew

def impute_airbnb_data(df):
    # ---------- NUMERIC IMPUTATION FUNCTION ----------
    def impute_numeric(column, group_cols=None, skew_thresh=0.5):
        if column not in df.columns or df[column].isna().sum() == 0:
            return
        if group_cols:
            df[column] = df.groupby(group_cols)[column].transform(
                lambda x: x.fillna(x.median() if abs(skew(x.dropna())) > skew_thresh else x.mean())
            )
        col_data = df[column].dropna()
        if not col_data.empty:
            value = col_data.median() if abs(skew(col_data)) > skew_thresh else col_data.mean()
            df[column] = df[column].fillna(value)

    # ---------- CATEGORICAL IMPUTATION FUNCTION ----------
    def impute_categorical(column, group_cols=None, fallback="NA"):
        if column not in df.columns or df[column].isna().sum() == 0:
            return
        if group_cols:
            df[column] = df.groupby(group_cols)[column].transform(
                lambda x: x.fillna(x.mode()[0] if not x.mode().empty else fallback)
            )
        mode_val = df[column].mode()
        df[column] = df[column].fillna(mode_val[0] if not mode_val.empty else fallback)

    # ---------- LOCATION-BASED ----------
    impute_categorical("host_location", group_cols=["host_is_superhost"])
    impute_categorical("host_neighbourhood", group_cols=["host_location"])
    impute_categorical("neighbourhood", group_cols=["neighbourhood_cleansed"])

    # ---------- HANDLE NEIGHBOURHOOD CLEANSED ----------
    if "neighbourhood_cleansed" in df.columns and df["neighbourhood_cleansed"].isna().sum() > 0:
        mode_val = df["neighbourhood_cleansed"].mode()
        if not mode_val.empty:
            df["neighbourhood_cleansed"] = df["neighbourhood_cleansed"].fillna(mode_val[0])

    # ---------- RATING COLUMNS ----------
    rating_cols = [
        "review_scores_rating", "review_scores_accuracy", "review_scores_cleanliness",
        "review_scores_checkin", "review_scores_communication", "review_scores_location",
        "review_scores_value"
    ]
    for col in rating_cols:
        impute_numeric(col, group_cols=["neighbourhood_cleansed"])

    # ---------- DATE COLUMNS ----------
    for col in ["first_review", "last_review"]:
        if col in df.columns and df[col].isna().sum() > 0:
            df[col] = pd.to_datetime(df[col], errors='coerce')
            median_val = df[col].dropna().median()
            df[col] = df[col].fillna(median_val)

    # ---------- HOST STATS ----------
    impute_numeric("reviews_per_month", group_cols=["host_is_superhost"])
    impute_numeric("host_response_rate", group_cols=["host_response_time"])
    impute_numeric("host_acceptance_rate", group_cols=["host_is_superhost"])
    impute_numeric("availability_365", group_cols=["availability_90"])

    # If host_response_time still missing, impute from superhost
    impute_categorical("host_response_time", group_cols=["host_is_superhost"])
    
    # If host_is_superhost is still missing, impute with mode
    impute_categorical("host_is_superhost")

    # ---------- ROOM / PROPERTY DETAILS ----------
    impute_categorical("property_type", group_cols=["room_type"])
    impute_categorical("room_type", group_cols=["property_type"])
    impute_numeric("bedrooms", group_cols=["accommodates", "room_type"])
    impute_numeric("beds", group_cols=["bedrooms", "accommodates"])

    # ---------- BATHROOM ----------
    impute_numeric("bathroom_qty", group_cols=["bedrooms", "accommodates", "room_type"])
    impute_categorical("bathroom_type", group_cols=["room_type"])

    # ---------- NIGHT LIMITS ----------
    impute_numeric("minimum_minimum_nights", group_cols=["minimum_nights"])
    impute_numeric("maximum_minimum_nights", group_cols=["minimum_nights"])
    impute_numeric("minimum_maximum_nights", group_cols=["maximum_nights"])
    impute_numeric("maximum_maximum_nights", group_cols=["maximum_nights"])
    impute_numeric("minimum_nights_avg_ntm", group_cols=["minimum_nights"])
    impute_numeric("maximum_nights_avg_ntm", group_cols=["maximum_nights"])

    # ---------- TEXT FIELDS ----------
    for col in ["description", "host_about", "host_neighbourhood", "neighborhood_overview"]:
        if col in df.columns:
            df[col] = df[col].fillna("NA")

    return df
```

```python
from sklearn.preprocessing import LabelEncoder

def encode_categorical_columns(df, low_cardinality, high_cardinality):
    """
    Encodes categorical features in a DataFrame:
    - For high-cardinality columns: top-5 encoding + 'other' = 0
    - For low-cardinality columns: label encoding
    
    Parameters:
    - df: DataFrame to encode
    - low_cardinality: list of column names with low cardinality
    - high_cardinality: list of column names with high cardinality
    
    Returns:
    - df: encoded DataFrame (modified in-place)
    """
    df = df.copy()

    # Encode high-cardinality columns
    for col in high_cardinality:
        if col in df.columns:
            df[col] = df[col].astype(str)
            df[col] = df[col].apply(lambda x: 'other' if ',' in x else x)
            top5 = df[col].value_counts().nlargest(5).index.tolist()
            mapping = {'other': 0}
            for i, val in enumerate(top5, start=1):
                mapping[val] = i
            df[col] = df[col].apply(lambda x: mapping.get(x, 0))

    # Label encode low-cardinality columns
    le = LabelEncoder()
    for col in low_cardinality:
        if col in df.columns:
            df[col] = le.fit_transform(df[col].astype(str))

    return df
```

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from scipy.stats import zscore
import numpy as np

# ---------------- Prepare data ----------------
X = airbnb_train_encoded.select_dtypes(include=[np.number]).drop(columns=['price'])
y = airbnb_train['log_price']

# ---------------- Preprocessing pipeline ----------------
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), X.columns)
])

# ---------------- Ridge model to estimate residuals ----------------
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', Ridge(alpha=1.0))
])

pipeline.fit(X, y)
y_pred = pipeline.predict(X)
residuals = y - y_pred

# ---------------- Z-score on residuals ----------------
z_scores = zscore(residuals)
threshold = 3  # you can reduce this to 2.5 or 2.0 if you want more aggressive filtering

# ---------------- Filter inliers ----------------
inliers_mask = np.abs(z_scores) < threshold
airbnb_train_supervised_cleaned = airbnb_train_encoded[inliers_mask].reset_index(drop=True)

# ---------------- Summary ----------------
print(f"Original data shape: {airbnb_train.shape}")
print(f"Cleaned data shape: {airbnb_train_supervised_cleaned.shape}")
print(f"Outliers removed: {(~inliers_mask).sum()}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, StackingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.base import clone
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import shap

# ---------------- Add KMeans Cluster Feature ----------------
coords_train = airbnb_train_supervised_cleaned[['latitude', 'longitude']]
coords_test = airbnb_test[['latitude', 'longitude']]

sil_scores = []
cluster_range = range(2, 11)
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(coords_train)
    sil_scores.append(silhouette_score(coords_train, labels))

optimal_k = cluster_range[np.argmax(sil_scores)]
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
airbnb_train_supervised_cleaned['location_cluster'] = kmeans.fit_predict(coords_train)
airbnb_test['location_cluster'] = kmeans.predict(coords_test)

# ---------------- Prepare Data ----------------
X_train_full = airbnb_train_supervised_cleaned.select_dtypes(include=[np.number]).drop(columns=['price', 'log_price'])
y_train = airbnb_train_supervised_cleaned['log_price']
X_test_full = airbnb_test.select_dtypes(include=[np.number])

# ---------------- SHAP Feature Selection ----------------
print(" Running SHAP for feature selection...")
shap_model = XGBRegressor(n_estimators=100, max_depth=5, learning_rate=0.1, random_state=42)
shap_model.fit(X_train_full, y_train)
explainer = shap.Explainer(shap_model, X_train_full)
shap_values = explainer(X_train_full)
shap_importance = np.abs(shap_values.values).mean(axis=0)
shap_summary = pd.DataFrame({
    'feature': X_train_full.columns,
    'importance': shap_importance
}).sort_values(by='importance', ascending=False)

top_features = shap_summary.head(21)['feature'].tolist()
X_train = X_train_full[top_features]
X_test = X_test_full[top_features]

# ---------------- CV Setup ----------------
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# ---------------- Define Models and Hyperparameters ----------------
model_grids = {
    'ridge': (Ridge(), {'regressor__alpha': [0.1, 1.0, 10.0]}),
    'lasso': (Lasso(), {'regressor__alpha': [0.01, 0.1, 1.0]}),
    'enet': (ElasticNet(), {'regressor__alpha': [0.01, 0.1], 'regressor__l1_ratio': [0.3, 0.5, 0.7]}),
    'rf': (RandomForestRegressor(), {'regressor__n_estimators': [100], 'regressor__max_depth': [10, 20]}),
    'xgb': (XGBRegressor(), {'regressor__n_estimators': [100], 'regressor__max_depth': [3, 6]}),
    'lgbm': (LGBMRegressor(), {'regressor__n_estimators': [100], 'regressor__num_leaves': [31, 50]}),
    'catboost': (CatBoostRegressor(verbose=0), {'regressor__iterations': [100], 'regressor__depth': [4, 6]}),
    'svr': (SVR(), {'regressor__C': [1.0, 10.0], 'regressor__epsilon': [0.1, 0.2]}),
    'hgbm': (HistGradientBoostingRegressor(random_state=42), {'regressor__learning_rate': [0.05, 0.1], 'regressor__max_iter': [100, 200]}),
    'lr': (LinearRegression(), {})
}

best_estimators = {}
model_scores = {}

# ---------------- Model Tuning ----------------
for name, (model, grid) in model_grids.items():
    print(f" Tuning: {name.upper()}")
    pipe = Pipeline([
        ('preprocessor', StandardScaler()),
        ('regressor', model)
    ])
    search = GridSearchCV(pipe, grid, cv=kfold, scoring='neg_root_mean_squared_error', n_jobs=-1)
    search.fit(X_train, y_train)
    best_model = search.best_estimator_.named_steps['regressor']
    score = -search.best_score_
    best_estimators[name] = (name, best_model, top_features)
    model_scores[name] = score
    print(f" {name.upper()} - Best RMSE: {score:.4f}, Params: {search.best_params_}")

# ---------------- Use Top 5 Models ----------------
top_5_models = sorted(model_scores.items(), key=lambda x: x[1])[:5]
stacking_estimators = [(name, clone(best_estimators[name][1])) for name, _ in top_5_models]
meta_model = best_estimators['enet'][1]

stacking_model = StackingRegressor(
    estimators=stacking_estimators,
    final_estimator=meta_model,
    passthrough=False,
    n_jobs=-1
)

# ---------------- Final Train and Predict ----------------
final_pipeline = Pipeline([
    ('preprocessor', StandardScaler()),
    ('stacking', stacking_model)
])

print(" Training stacking ensemble with top 5 models...")
final_pipeline.fit(X_train, y_train)

preds_log = final_pipeline.predict(X_test)
preds_price = np.expm1(preds_log)

# ---------------- Export Submission ----------------
submission = pd.DataFrame({
    'ID': airbnb_test['ID'],
    'price': preds_price
})
submission.to_csv("stacking_top5.csv", index=False)
print(" Submission saved as stacking_top5.csv")
```
